```python
# save all slices

# Define the range of patient numbers to loop through
start_num = 1
end_num = 42

for num in range(start_num, end_num + 1):
  # Format the patient number with leading zeros
  num_str = f"{num:04d}"
  print(num_str)

  # Define full paths
  # num = "0001"
  n = "n4"
  hdr_path = f"oasis-1-dataset-13635/oasis_raw/disc1/OAS1_{num_str}_MR1/PROCESSED/MPRAGE/T88_111/OAS1_{num_str}_MR1_mpr_{n}_anon_111_t88_masked_gfc.hdr"
  img_path = f"oasis-1-dataset-13635/oasis_raw/disc1/OAS1_{num_str}_MR1/PROCESSED/MPRAGE/T88_111/OAS1_{num_str}_MR1_mpr_{n}_anon_111_t88_masked_gfc.img"

  # Download .hdr
  with fs.open(hdr_path, 'rb') as f:
      with open('mpr.hdr', 'wb') as out_file:
          out_file.write(f.read())

  # Download .img
  with fs.open(img_path, 'rb') as f:
      with open('mpr.img', 'wb') as out_file:
          out_file.write(f.read())

  import nibabel as nib
  import matplotlib.pyplot as plt
  import os
  from google.colab import drive

  # Load Analyze image
  img = nib.load('mpr.img')
  img_data = img.get_fdata()

  # # Mount Google Drive
  # drive.mount('/content/gdrive', force_remount=True)

  # Create output folder
  output_folder = '/content/gdrive/My Drive/OASIS_Slices_JPG/'
  os.makedirs(output_folder, exist_ok=True)

  # Extract original filename
  original_name = os.path.splitext(os.path.basename(hdr_path))[0]

  # Modified slice extraction - ALL SLICES
  num_slices = img.shape[2]  # Get total number of slices
  slices = img_data[:, :, :, 0]  # Take all slices in 3rd dimension

  # Save all slices
  for i in range(num_slices):
      filename = os.path.join(output_folder, f"{original_name}_slice{i:03d}.jpg")

      plt.imshow(slices[:, :, i].T, cmap='gray', origin='lower')
      # plt.title(f'Slice {i}')
      plt.axis('off')

      plt.savefig(filename,
                  bbox_inches='tight',
                  pad_inches=0,
                  dpi=150,
                  format='jpg',
                  pil_kwargs={'quality': 90, 'optimize': True})
      plt.close()

  print(f"Saved ALL {num_slices} JPG slices to:\n{output_folder}")
```

```python
# --- Plotting All Slices in One Figure ---
num_cols = 8  # Number of columns in the grid
num_rows = (num_slices + num_cols - 1) // num_cols  # Calculate number of rows

fig, axes = plt.subplots(num_rows, num_cols, figsize=(24, 3 * num_rows))  # Adjust figure size

# Flatten the axes array to make indexing easier
axes = axes.flatten()

# Plot each slice
for i in range(num_slices):
    ax = axes[i]  # Get the axis for this slice
    ax.imshow(slices[:, :, i].T, cmap='gray', origin='lower')
    ax.set_title(f'Slice {i}')
    ax.axis('off')

# Remove any unused subplots
for i in range(num_slices, num_rows * num_cols):
    fig.delaxes(axes[i])

plt.tight_layout()

# Save the combined plot
combined_filename = os.path.join(output_folder, f"{original_name}_all_slices.jpg")
plt.savefig(combined_filename, dpi=150)
plt.close()

print(f"Saved combined JPG slices to:\n{output_folder}")
```
